```python
df_numerical_features = df2.select_dtypes(include='number')
df_numerical_features.columns
```

```python
# Assuming 'df_numerical_features' is a DataFrame with numerical columns.
# We will create a sample DataFrame to simulate this scenario.

import numpy as np
import matplotlib.pyplot as plt


# Sample data creation
np.random.seed(0)

# Plotting Histogram with KDE and Boxplot for each numerical feature
fig, axes = plt.subplots(len(df_numerical_features.columns), 2, figsize=(12, 5 * len(df_numerical_features.columns)))

for i, col in enumerate(df_numerical_features.columns):
    # Histogram with KDE
    df_numerical_features[col].plot(kind='hist', density=True, ax=axes[i, 0], bins=20, alpha=0.5, color='blue')
    df_numerical_features[col].plot(kind='kde', ax=axes[i, 0], color='red')
    axes[i, 0].set_title(f'{col} - Histogram with KDE')

    # Boxplot
    df_numerical_features.boxplot(column=col, ax=axes[i, 1])
    axes[i, 1].set_title(f'{col} - Boxplot')

plt.tight_layout()
plt.show()
```

```python
#NOTES#####
# total pay bebefits have outliers  in range <500000
#total pay  have outliers  in range between 400000 and 500000
#benefits have outliers in range between 90000 and 100000
#otherpay have 3 outliers (250000,340000,400000)
#over time pay have outliers in range (180000,250000)
```

```python
import seaborn as sns
corr = df_numerical_features.corr()

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Heatmap of Numerical Features')
plt.show()
```

```python
#BasePay is highly  correlated with Benefits and totalpay andtotalbaybenefits
#Benefits is highly  correlated with BasePay and totalpay andtotalbaybenefits
#TotalPay is highly  correlated with BasePay and Benefits andtotalbaybenefits
#TotalPayBenefits is highly  correlated with BasePay and totalpay and benifits
```

```python
# scatter

n_features = df_numerical_features.shape[1] - 1
fig, axs = plt.subplots(n_features, figsize=(10, n_features * 5))

for i, feature in enumerate(df_numerical_features.columns[:-1]):
    axs[i].scatter(df_numerical_features[feature], df_numerical_features['TotalPay'])
    axs[i].set_xlabel(feature)
    axs[i].set_ylabel('TotalPay')
    axs[i].set_title(f'Scatter plot of {feature} vs TotalPay')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called 'df' with a column called 'job_title'
job_counts = df['JobTitle'].value_counts()

# Select the top 3 most common job titles
top_3_jobs = job_counts[:3]

# Create a bar plot of the top 3 job titles
plt.bar(top_3_jobs.index, top_3_jobs.values)
plt.title('Top 3 Most Common Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame called 'df' with columns 'job' and 'salary'
group_salaries = df.groupby('JobTitle')['TotalPay'].mean()

# Create a bar plot of the average salaries by job
plt.bar(group_salaries.index, group_salaries.values)
plt.title('Average Salaries by job')
plt.xlabel('job')
plt.ylabel('Average Salary')
plt.show()
```
